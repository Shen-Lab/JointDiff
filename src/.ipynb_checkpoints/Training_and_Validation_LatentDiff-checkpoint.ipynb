```python
def ave_plot(iter_list, val_list, interval = 100, label = None, with_log = False):
    l = len(iter_list)
    point_num = l // interval

    if with_log:
         val_list = [np.log(i) for i in  val_list]
    
    new_iter = []
    new_val = []

    for i in range(point_num):
        new_iter.append((i+1) * interval)
        new_val.append(np.mean(val_list[i * interval : (i+1) * interval]))

    plt.plot(new_iter, new_val, label = label)
```

```python
plt.figure(figsize=(12,4))

plt.subplot(1,2,1)
ave_plot(loss_dict['iter'], loss_dict['loss'])
plt.ylabel('Overall Loss', fontsize = 12)
plt.xlabel('Iteration', fontsize = 12)

plt.subplot(1,2,2)
for key in ['esm_feat_recover', 'esm_pair_recover', 'mpnn_feat_recover', 'mpnn_pair_recover', 
            'resi-wise-simi', 'pair-wise-simi', 'dist_pred']:
    ave_plot(loss_dict['iter'], loss_dict[key], label = key)

plt.ylabel('Loss', fontsize = 12)
plt.xlabel('Iteration', fontsize = 12)
plt.legend(fontsize = 10)

plt.show()
```

```python
def loss_summary(
    loss_dict, 
    keys = [
        'esm_feat_recover', 
        'esm_pair_recover', 
        'mpnn_feat_recover', 
        'mpnn_pair_recover', 
        'dist_pred'
    ]
):
    return list(
        np.array([loss_dict[k] for k in keys]).sum(axis=0)
    )
```
